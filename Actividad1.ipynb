```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

```python
im_roja  = imread('img_rojo.jpeg')
im_verde = imread('img_verde.jpeg')
im_azul = imread('img_azul.jpeg')
```

```python
alto, ancho, capas = im_roja.shape
print("El tamaño de la imagen es: ")
print("Alto: ", alto, "px\nAncho:", ancho, "px\nCapas:", capas, "px")
```

```python
# Mostrando la imagen con filtro rojo, verde y azul
plt.figure(figsize=(10, 4))

# Subgráfico 1: Imagen Roja
plt.subplot(1, 3, 1)
plt.imshow(im_roja, cmap='gray')
plt.title('Imagen Roja Original')

# Subgráfico 2: Imagen Verde
plt.subplot(1, 3, 2)
plt.imshow(im_verde, cmap='gray')
plt.title('Imagen Verde Original')

# Subgráfico 3: Imagen Azul
plt.subplot(1, 3, 3)
plt.imshow(im_azul, cmap='gray')
plt.title('Imagen Azul Original')

plt.show()
```

```python
# tomamos las partes de cada imagen filtrada recordando el orden RGB
parte_roja  = im_roja[:,:,0]  # R
parte_verde = im_verde[:,:,1] # G
parte_azul  = im_azul[:,:,2]  # B
```

```python
plt.figure(figsize=(10, 4))

# Subgráfico 1: Parte Roja
plt.subplot(1, 3, 1)
plt.imshow(parte_roja, cmap='gray')
plt.title('Parte Roja')

# Subgráfico 2: Parte Verde
plt.subplot(1, 3, 2)
plt.imshow(parte_verde, cmap='gray')
plt.title('Parte Verde')

# Subgráfico 3: Parte Azul
plt.subplot(1, 3, 3)
plt.imshow(parte_azul, cmap='gray')
plt.title('Parte Azul')

plt.show()
```

```python
imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
```

```python
# mostremos la imagen completa...
plt.imshow(imagen_completa)
```
